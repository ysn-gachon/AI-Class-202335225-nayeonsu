```python
import numpy as np

import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/titanic.csv')

from sklearn.preprocessing import LabelEncoder







le = LabelEncoder()






df.head()
```

```python
# 각 컬럼별로 빈 데이터(NaN)가 몇 개인지 출력
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())

# 가장 빈도가 높은 값(최빈값)으로 채우기
mode_val = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(mode_val)
```

```python
# Name 컬럼 인코딩

df['Name_encoded'] = le.fit_transform(df['Name'])
df['Sex_encoded'] = le.fit_transform(df['Sex'])
df['Ticket_encoded'] = le.fit_transform(df['Ticket'])
df['Embarked_encoded'] = le.fit_transform(df['Embarked'])
```

```python
y = df['Survived']
print(y.value_counts())
```

```python
X = df.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin', 'Embarked', 'Sex'], axis=1)


X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(X_train, y_train)

pred_dt = clf_dt.predict(X_test)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print ("\n--- Random Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```
